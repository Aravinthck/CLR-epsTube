```python
def Blobs(N,D,K,box,seed):


    generator = np.random.RandomState(seed)
    regParam = np.round(generator.uniform(-1,1, size=(K, D+1)),2)
    print(regParam)
    # box = 1
    generator = np.random.RandomState(seed)
    centers = generator.uniform(0, box, size=(K, D))
    ratio_std_x = 0.07
    ratio_std_y = 0.008

    X, y = make_blobs(n_samples=N, centers=centers, n_features=D, cluster_std = ratio_std_x*box,
                        random_state=seed)
    # print(centers)
    cols = [ 'X' + str(i+1) for i in range(D) ]
    df_data = pd.DataFrame(X, columns = cols)

    for k in range(K):
        generator = np.random.RandomState(seed)
        df_data.loc[y==k,'Y'] = X[y == k] @ regParam[k,:-1] + regParam[k,-1] + generator.normal(0,ratio_std_y*box,X[y==k].shape[0])

    df_data.plot('X1', 'Y', kind = 'scatter', s = 20)
    plt.show()

    return df_data, X,y
```

```python
N=100
K=3
D=1
box = 1
seed =123

df_data, X,y = Blobs(N,D,K,box,seed)
```
